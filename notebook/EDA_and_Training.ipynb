```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('C:/projek web penyakit/data/dataset.csv')
df_desc = pd.read_csv('C:/projek web penyakit/data/symptom_Description.csv')
df_precaution = pd.read_csv('C:/projek web penyakit/data/symptom_precaution.csv')
df_severity = pd.read_csv('C:/projek web penyakit/data/Symptom-severity.csv')


df.head()
```

```python
# Cek missing values
print("=== Missing Values ===")
print(df.isnull().sum())

print("\n=== Jumlah penyakit unik ===")
print(df['Disease'].nunique())

print("\n=== Distribusi penyakit ===")
print(df['Disease'].value_counts())
```

```python
# Hapus spasi tersembunyi di nama kolom & nilai
df.columns = df.columns.str.strip()
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Ambil semua kolom gejala
symptom_cols = [col for col in df.columns if 'Symptom' in col]
print("Kolom gejala:", symptom_cols)

# Kumpulkan semua gejala unik
all_symptoms = pd.unique(df[symptom_cols].values.ravel())
all_symptoms = [s for s in all_symptoms if str(s) != 'nan']
all_symptoms = sorted(list(set([s.strip() for s in all_symptoms])))

print(f"\nTotal gejala unik: {len(all_symptoms)}")
```

```python
# Buat dataframe baru dengan one-hot encoding
encoded_data = []

for _, row in df.iterrows():
    symptoms_in_row = [row[col] for col in symptom_cols if str(row[col]) != 'nan']
    symptoms_in_row = [s.strip() for s in symptoms_in_row]
    
    encoded_row = {symptom: 1 if symptom in symptoms_in_row else 0 
                   for symptom in all_symptoms}
    encoded_row['Disease'] = row['Disease']
    encoded_data.append(encoded_row)

df_encoded = pd.DataFrame(encoded_data)
print("Shape setelah encoding:", df_encoded.shape)
df_encoded.head()
```

```python
# Simpan untuk dipakai di step modeling
df_encoded.to_csv('C:/projek web penyakit/data/df_encoded.csv', index=False)

# Simpan list gejala
import json
with open('C:/projek web penyakit/data/symptoms_list.json', 'w') as f:
    json.dump(all_symptoms, f)

print(" Data berhasil disimpan!")
print(f"- df_encoded.csv ({df_encoded.shape})")
print(f"- symptoms_list.json ({len(all_symptoms)} gejala)")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Pisah fitur dan target
X = df_encoded.drop('Disease', axis=1)
y = df_encoded['Disease']

# Encode label penyakit jadi angka
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"Training set : {X_train.shape}")
print(f"Testing set  : {X_test.shape}")
print(f"Jumlah kelas : {len(le.classes_)}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'SVM'          : SVC(probability=True, random_state=42),
    'XGBoost'      : XGBClassifier(random_state=42, eval_metric='mlogloss')
}

results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name}: {acc*100:.2f}%")
```

```python
import joblib

# Pilih model dengan akurasi tertinggi
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]

print(f"Model terbaik: {best_model_name} ({results[best_model_name]*100:.2f}%)")

# Simpan model dan label encoder
joblib.dump(best_model, 'C:/projek web penyakit/model/model.pkl')
joblib.dump(le, 'C:/projek web penyakit/model/label_encoder.pkl')

print(" Model berhasil disimpan!")
```

```python
import shap

# Buat explainer dari Random Forest
explainer = shap.TreeExplainer(best_model)

# Hitung SHAP values dari data test
shap_values = explainer.shap_values(X_test)

print(f" SHAP values berhasil dihitung!")
print(f" Shape SHAP values: {len(shap_values)} kelas")
```

```python
# Rata-rata SHAP values untuk semua kelas
shap_values_mean = np.abs(shap_values).mean(axis=0)

# Ambil top 20 fitur paling berpengaruh
feature_names = X_test.columns.tolist()
shap_df = pd.DataFrame({
    'Symptom': feature_names,
    'Importance': shap_values_mean.mean(axis=1)
}).sort_values('Importance', ascending=False).head(20)

fig = px.bar(shap_df,
             x='Importance', y='Symptom',
             orientation='h',
             title='Top 20 Gejala Paling Berpengaruh (SHAP)',
             color='Importance',
             color_continuous_scale='Purples')

fig.update_layout(height=600, yaxis={'categoryorder':'total ascending'})
fig.show()
```
